```python
%matplotlib inline
from matplotlib import colors
import numpy as np
import matplotlib.pyplot as plt
import operator

```

```python
def func(point):
    x1= point[0]
    x2= point[1]
    return (1-x1)**2+100*(x2-x1**2)**2
```

```python
def NM(v1,v2,v3):
    alpha=1
    beta=0.5
    gamma=2
    max_iter=5
    e=(1,1)
    for i in range(max_iter):
        values={v1:func(v1),v2:func(v2),v3:func(v3)}
        points= sorted(values.items(), key=lambda x:x[1])
        #print(points)
        b = points[0][0]
        g = points[1][0]
        w = points[2][0]
        mid = ((g[0]+b[0])/2,(g[1]+b[1])/2)
        #reflection
        #r= mid + alpha*(mid-w)
        r = (mid[0] + alpha *(mid[0]-w[0]),mid[1] + alpha *(mid[1]-w[1]))
        if(func(r)<func(g) and func(r)>func(b)):
            w=r
        #expansion
        if(func(r)<func(b)):
            e= mid + gamma*(r-mid)
            if(func(e)< func(r)):
                w=e
            else:
                w=r
        #contraction
        if(func(r)>func(g)):
            #c1= w+ 0.25*(r-w)
            c1 = (w[0] + 0.25 *(r[0]-w[0]),w[1] + 0.25 *(r[1]-w[1]))
            #c2= w+ 0.75*(r-w)
            c2 = (w[0] + 0.75 *(r[0]-w[0]),w[1] + 0.75 *(r[1]-w[1]))
            if(func(c1)<func(g) and func(c1)<func(c2)):
                w=c1
            elif(func(c2)<func(g) and func(c2)< func(c1)):
                w=c2
            else:
                #w= b+ beta*(b-w)
                w = (b[0] + beta *(b[0]-w[0]),b[1] + beta *(b[1]-w[1]))
                #g= b+ beta*(b-g)
                g = (b[0] + beta *(b[0]-g[0]),b[1] + beta *(b[1]-g[1]))
        #update points
        v1=w
        v2=g
        v3=b
    return b
                
```
